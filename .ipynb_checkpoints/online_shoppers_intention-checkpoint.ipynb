## Посчитаем Informational Value для каждого признака

```python
def calc_iv(df, feature, target ='Revenue'):
    df_woe = (pd.crosstab(df[feature],df[target],
                      normalize='columns')
             .assign(woe=lambda dfx: np.log(dfx[True] / dfx[False])))
    df_woe[(df_woe == np.inf) | (df_woe ==-np.inf)] = 0
    return np.sum(df_woe['woe']*(df_woe[True]-df_woe[False]))
    
def calc_all_IV(df):
    result = pd.DataFrame(columns=['feature', 'IV']) 
    for value in df.columns:
        if value != 'Revenue':
            result.loc[len(result.index)] = [value, calc_iv(df, value)]
        
    return result
```

```python
df_iv = df.copy()

calc_all_IV(df_iv).sort_values(by=['IV'], ascending = False)
```

### У 5 признаков Informational Value >0.3, что характеризует их высокую значимость. У 4 признаков 0.1<IV<0.3, что можно интерпретировать как средняя значимость

# Построим предсказательную модель

```python
#Переведем True->1, False->0

le = LabelEncoder()
df['Revenue'] = le.fit_transform(df['Revenue'])
df['Revenue'].value_counts()
```

```python
df
```

```python
#Удалим сильно коррелирующие переменные
df = df.drop(['BounceRates', 'ProductRelated', 'Informational', 'Administrative'], axis=1)
```

```python
from sklearn.model_selection import train_test_split

df = pd.get_dummies(df)
X = df.drop(['Revenue'], axis=1)
y = df['Revenue']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

### Данная выборка несбалансированна. Для борьбы с этим можно использовать oversampling с генерацией новых похожих значений

```python
from imblearn.over_sampling import SMOTE


smote = SMOTE(sampling_strategy='minority')
X_train, y_train = smote.fit_resample(X_train, y_train)
```

```python
from lazypredict.Supervised import LazyClassifier


clf = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)

models,predictions = clf.fit(X_train, X_test, y_train, y_test)

print(models)
```

### Для создания предсказательной модели будем использовать XGBClassifier. Чтобы подобрать наилучшие гиперпараметры воспользуемся GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
import xgboost as xgb


param_grid = {'max_depth':[5,6], 'n_estimators':[2000, 3000], 
              "learning_rate":[0.003, 0.001]}

gs= GridSearchCV(cv=5, error_score='raise',
                estimator=xgb.XGBClassifier(),
                n_jobs=-1,
                scoring='roc_auc',
                param_grid = param_grid                                                                        
                                       ) 
```

```python
gs.fit(X_train,y_train)
```

### Выведем результаты для наилучших параметров и наиболее значимые фичи

```python
print(gs.best_estimator_)
print('best score = ', gs.best_score_)
print('best params = ', gs.best_params_, '\n')

y_pred = gs.best_estimator_.predict(X_test)
print(sklearn.metrics.classification_report(y_test, y_pred))

plt.rcParams['figure.figsize'] = (16, 10)
xgb.plot_importance(gs.best_estimator_)
```

### Рассмотрим, как менялись ошибка и AUC в процессе обучения 

```python
params = {
 "model_params": gs.best_params_,
  "train_params": {
    "eval_metric": [
      "error",
      "auc"
    ]
  }
}

model = xgb.XGBClassifier(**params["model_params"], use_label_encoder=False)
model.fit(X_train, y_train, 
          eval_set=[(X_train, y_train), (X_test, y_test)], 
          **params["train_params"])

train_history = model.evals_result()
```

```python
plt.rcParams['font.size'] = 18
plt.rcParams['figure.figsize'] = (8, 8)

plt.title("AUC")
plt.plot(train_history["validation_0"]["auc"], label="training data")
plt.plot(train_history["validation_1"]["auc"], label="test data")
plt.legend()
plt.show()

plt.title("Error")
plt.plot(train_history["validation_0"]["error"], label="training data")
plt.plot(train_history["validation_1"]["error"], label="test data")
plt.legend()
plt.show()


y_pred_proba_train = model.predict_proba(X_train)
y_pred_proba_test = model.predict_proba(X_test)

fpr_test, tpr_test, _ = sklearn.metrics.roc_curve(y_test, y_pred_proba_test[:, 1])
fpr_train, tpr_train, _ = sklearn.metrics.roc_curve(y_train, y_pred_proba_train[:, 1])

auc_test = sklearn.metrics.auc(fpr_test, tpr_test)
auc_train = sklearn.metrics.auc(fpr_train, tpr_train)

plt.title(f"ROC curve, AUC=(test: {auc_test:.4f}, train: {auc_train:.4f})")
plt.plot(fpr_train, tpr_train, label="train data")
plt.plot(fpr_test, tpr_test, label="test data")
plt.legend()
plt.show()
```

## На графике AUC видно небольшое переобучение, начавшееся в районе 2250 обученных деревьев. Установив ограничение n_estimators = 2250 можно получить зачение AUC на тестовой выборке равным 0.93283 против текущих 0.93261. Отличие возникает только в 4 знаке после запятой
